```python
#Ordinary (n,1) cable
n=5
signs=[1 for i in range(2*n)]
over_strands=n*[[1,0],[0,0]]
knot=Link_Component(2,signs,over_strands)

pb_signs=[1 for i in range(n)]
pb_over_strands=[[0,2*i+1] for i in range(n)]
pb=Link_Component(2,pb_signs,pb_over_strands)

diagram=Link_Diagram(2,0)
diagram.add_component(knot)
diagram.add_component(pb)

cover=Cyclic_Cover(n,diagram)

```

```python
for i in range(1,n+1):
    print("2-chain bounding lift", i)
    chain=cover.two_chain(1,i)
    for x in range(len(chain)):
        print("x_",x // n,"^",x%n +1,"=",chain[x] )
        if (x+1)%n==0:
            print()
    for j in range(1,n+1):
        if i!=j:
            print("lift ",i, "with lift ",j)
            print(cover.linking_number(1,i,1,j))
    print()
```

```python
#(n,1) cable with rightmost k crossings changed, degree q cover.  Note q must divide n.

def C_n_1_k(n,k,q):

    signs=k*[-1]+(2*n-k)*[1]
    over_strands=[[0,2*k-i] for i in range(k)]+ (k+1)*[[1,0]]+(n-k-1)*[[0,0],[1,0]]+[[0,2*n+i] for i in range(2*k)]+[[0,0]]
    knot=Link_Component(2,signs,over_strands)
    print("Knot signs", signs)
    print("Knot overstrands", over_strands)

    pb_signs=[1 for i in range(n)]
    pb_over_strands=[[0,i] for i in range(k+1,2*k+1)]+[[0,2*i+1] for i in range(k,n)]
    pb=Link_Component(2,pb_signs,pb_over_strands)

    diagram=Link_Diagram(2,0)
    diagram.add_component(knot)
    diagram.add_component(pb)

    cover=Cyclic_Cover(q,diagram)
    
    for j in range(1,q+1):
        print("2-chain bounding lift", j)
        chain=cover.two_chain(1,j)
        for x in range(len(chain)):
            print("x_",x // n,"^",x%n +1,"=",chain[x] )
            if (x+1)%n==0:
                print()
    
    
    for j in range(2,q+1):
        
#         print("2-chain bounding lift", j)
#         chain=cover.two_chain(1,j)
#         for x in range(len(chain)):
#             print("x_",x // n,"^",x%n +1,"=",chain[x] )
#             if (x+1)%n==0:
#                 print()
        
        print("lift ",1, "with lift ",j)
        print(cover.linking_number(1,1,1,j))
```

```python
#Examples where degree of cover is n, standard cables

for n in range(6,7):
    print("-------------")
    print("n=",n)
    print("k=",0)
    C_n_1_k(n,0,n)
    print("-------------")
```

```python
#Examples where degree of cover is n, change k rightmost crossings

for n in range(3,4):
    for k in range(n):
        print("-------------")
        print("n=",n)
        print("k=",k)
        C_n_1_k(n,k,n)
        print("-------------")
```
